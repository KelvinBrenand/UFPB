```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```
df1 = pd.read_csv('graph2.csv')
df1
```

```
df1.loc[0] = df1.loc[0]/1000000
df1.loc[1] = df1.loc[1]/1000000
df1.loc[2] = df1.loc[2]/1000000
df1.loc[3] = df1.loc[3]/1000000
df1
```

```
plt.figure(figsize=(20,10))
row1 = df1.iloc[0]
row1.plot(label = 'Sequential version', kind='line', )
row2 = df1.iloc[1]
row2.plot(label = 'Parallel version with 2 threads', kind='line')
row3 = df1.iloc[2]
row3.plot(label = 'Parallel version with 3 threads', kind='line')
row4 = df1.iloc[3]
row4.plot(label = 'Parallel version with 4 threads', kind='line')
plt.ylabel('Time (seconds)')
plt.xlabel('N size')
plt.title("Parallel Programming")
plt.legend(loc='best')
#plt.axis([1, 6, 0, 1])
plt.show()
```
